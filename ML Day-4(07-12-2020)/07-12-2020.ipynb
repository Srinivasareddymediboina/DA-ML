````python
input_features=data[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
       'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'radius_se', 'perimeter_se',
       'area_se', 'concave points_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
````

````python
input_features
````

````python
output_target=data['diagnosis']
````

````python
output_target
````

````python
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te=train_test_split(input_features,
                                    output_target,test_size=0.3,
                                    random_state=2)
````

````python
# import the required algorithum
from sklearn.linear_model import LogisticRegression
````

````python
log=LogisticRegression()
````

````python
#Train the data using fit method
log.fit(x_tr,y_tr)
````

````python
pred_values=log.predict(x_te)
````

````python
# Evalute the model
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_te,pred_values)
````

- Change the random_state and test_size parametes values and check the accuracy

````python
confusion_matrix(y_te,pred_values)
````
