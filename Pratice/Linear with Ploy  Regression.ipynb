```python
import pandas as pd
```

```python
df=pd.read_csv('palcement.csv')
df.head()
```

```python
df.info()
```

```python
import seaborn as sns
```

```python
sns.pairplot(df)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['year'],df['placement data'])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df[['year']]
y=df['placement data']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=5,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
plt.scatter(df['year'],df['placement data'])
plt.plot(df['year'],model.predict(df[['year']]),color='r')
```

### Applying the Polynomial Features

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
ploy=PolynomialFeatures(degree=10)
```

```python
poly_train=ploy.fit_transform(x_train)
pd.DataFrame(poly_train)
```

```python
ploy_test=ploy.fit_transform(x_test)
ploy_test
```

```python
model
```

```python
model.fit(poly_train,y_train)
```

```python
model.score(poly_train,y_train)
```

```python
model.score(ploy_test,y_test)
```

```python
plt.scatter(df['year'],df['placement data'])
plt.plot(df['year'],model.predict(ploy.transform(df[['year']])),color='r')
```
