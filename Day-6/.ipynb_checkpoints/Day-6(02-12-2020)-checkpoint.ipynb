```python
import matplotlib.pyplot as plt
```

#### pie Chart
- plt.pie()

```python
sports=['cricket','volly','chess','tennies','footbal']
score=[80,90,98,58,69]
plt.pie(score,labels=sports,autopct='%1.1f%%',
        colors=['gold','yellow','m','pink','blue'],
        startangle=90,explode=(0,0,0,0,0.5),shadow=True)
plt.show()
```

```python
help(plt.pie)
```

#### Area Plot
- plt.stackplot()

```python
mng=[1,2,3,4]
night=[1,4,8,9]
evng=[3,5,6,9]
plt.stackplot(mng,night,evng,colors=['m','gold'])
```

### Builtin Datasets

```python

sns.get_dataset_names()
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris.shape
```

### Distribution plot
- It plots the univariantdistribution of data
- sns.distplot()

```python
sns.distplot(iris['sepal_length'])
```

```python
sns.set_style('whitegrid')
sns.distplot(iris['sepal_length'],kde=False)
```

### Categorical Plot
- by default it returns a scatter plot
- sns.catplot()

```python
sns.catplot(x='species',y='sepal_length',data=iris)
```

```python
# It displays only for numarical columns
sns.catplot(data=iris)
```

```python
sns.catplot(x='species',y='sepal_length',data=iris,palette='Blues')
```

- Boxplot

```python
sns.catplot(x='species',y='sepal_length',data=iris,kind='box')
```

```python
sns.catplot(data=iris,kind='box',orient='h',palette='husl')
```

- violin plot
    - it is very simiar to box plot

```python
sns.catplot(x='species',y='sepal_length',data=iris,kind='violin',
           palette='husl')
```

- strip plot
    - it is represented the data in sorted order

```python
sns.catplot(x='species',y='petal_length', data=iris,kind='strip')
```

- swarm plot
    - Also known as 'bees plot'
    - To Avoids Overlapping
    

```python
sns.catplot(x='species',y='petal_length',data=iris,kind='swarm')
```

### Joint Plot
- Combination of 2 plots, by default it conatains scatter and his plots
- sns.jointplot()

```python
iris.head()
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,kind='hex',color='gold')
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',
              data=iris,kind='kde',color='m')
```

### Pair plot
- its builds a histogram and scatter plot for pair wise relationship
- sns.pairplot()

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

```python
sns.pairplot(iris,vars=['sepal_length','sepal_width'])
```

```python
sns.pairplot(iris,hue='species',palette='husl')
```

```python
sns.pairplot(iris,hue='species',markers=['h','o','D'])
```

### Heatmaps
- Heatmap displays matrix format like graph,where all the cells are filled with colors based on values.
- sns.heatmap()

```python
t=sns.load_dataset('tips')
t.head()
```

```python
t.isnull().sum().sum()
```

```python
t.iloc[:,[0,1,-1]].head()
```

```python
help(sns.heatmap)
```

```python
sns.heatmap(t.iloc[:,[0,1,-1]].head())
```

```python
sns.heatmap(t.iloc[:,[0,1,-1]].head(),annot=True)
```

```python
t.corr()
```

```python
sns.heatmap(t.corr(),annot=True)
```

```python
sns.heatmap(t.corr(),annot=True,cmap='Reds')
```
