### Standard Scaler
- if data is normally distubuted , then we can apply standard scaler,
- it scales the data zero mean unit variance.
    - [X(I)-mean(x)]/std(x)

```python
df=pd.read_csv('Adv.csv')
df.head()
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df)
plt.show()
```

### Subplots

```python
fig, (ax1,ax2)=plt.subplots(1,2,figsize=(9,5))
```

```python
fig, (ax1,ax2)=plt.subplots(1,2,figsize=(10,5))

ax1.plot(x,x,color='green',label='Line')
ax1.plot(x,x**2,color='red',label='LinerLine')
ax2.plot(x,x**2, color='m')
ax1.set_xlabel('X_Axis')
ax1.set_ylabel('Y_Axis')
ax1.set_title('Liner Plot')
ax1.legend()
ax1.grid()

ax2.plot(x,x**2, color='yellow',label='lines')
ax2.plot(x,x**3, color='green',label='lines')
ax2.set_xlabel('X_Axis')
ax2.set_ylabel('Y_Axis')
ax2.set_title('Regular Plot')
ax2.legend()
ax2.grid()
```

### Boxplot
- it gives the summary of numarical data throught quantiles
- min
- IQR
    - Lower Quartile
    - Medium
    - Upper Quartile
- Max
- plt.boxplot()

```python
x=np.random.randint(2000,25000,1000)
x.max()
```

```python
plt.boxplot(x)
```

```python
help(plt.boxplot)
```

```python
plt.boxplot(x,notch=True,vert=False)
```

```python
games=['pubg','freefire','ludo']
ludo=np.random.randint(100,1000,50)
pubg=np.random.randint(150,5000,100)
freefire=np.random.randint(1000,5000,70)

plt.boxplot([ludo,freefire,pubg],labels=games,patch_artist=True,
            boxprops=dict(facecolor='gold',color='blue'))
plt.show()
```
