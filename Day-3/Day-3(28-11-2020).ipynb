### Replace Missing Values
- Replace missing values either with '0' or mean() value, if that data is numerical
- Replacemissing values with 'most occured object value ' or any object like 'no_name,None...'
    - fillna()

```python
data2=market.copy()
```

```python
data2.head(5)
```

```python
data2.shape
```

```python
data2.isnull().sum()
```

```python
data2['Product_Base_Margin'].mean()
```

```python
data2['Product_Base_Margin']=data2['Product_Base_Margin'].fillna(data2['Product_Base_Margin'].mean())
```
